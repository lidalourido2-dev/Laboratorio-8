```python
import random
```

```python
def generar_numero():
    """Genera un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)
```

Ahora implementaremos la lógica principal del juego: el bucle `while`, el manejo de entradas y las pistas para el jugador.

```python
numero_secreto = generar_numero()
intentos = 0

print("¡Bienvenido al juego de adivinar el número!")
print("Estoy pensando en un número entre 1 y 100.")

while True:
    try:
        adivinanza = int(input("Adivina el número: "))
        intentos += 1

        if adivinanza < numero_secreto:
            print("¡Demasiado bajo! Intenta de nuevo.")
        elif adivinanza > numero_secreto:
            print("¡Demasiado alto! Intenta de nuevo.")
        else:
            print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
            break
    except ValueError:
        print("Eso no es un número válido. Por favor, introduce un número.")
```

Ahora, implementaremos el menú principal de la calculadora, un bucle para seleccionar operaciones y el manejo de errores.

### Calculadora de Operaciones Básicas

```python
# 1. Definir funciones para cada operación
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ValueError("Error: No se puede dividir por cero.")
    return a / b
```

Ahora, implementaremos el menú principal de la calculadora con un bucle `while`, el manejo de entradas y errores usando `try-except`, y las opciones para continuar o salir.

```python
# 2. Crear un menú de opciones usando un bucle while
while True:
    print("\n--- Calculadora de Operaciones Básicas ---")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Salir")

    opcion = input("Selecciona una opción (1-5): ").strip()

    # 6. Implementar una opción para salir del programa
    if opcion == '5':
        print("¡Gracias por usar la calculadora! Hasta luego.")
        break

    # 3. Utilizar if-elif-else para seleccionar la operación
    if opcion in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Introduce el primer número: "))
            num2 = float(input("Introduce el segundo número: "))

            if opcion == '1':
                resultado = suma(num1, num2)
                print(f"Resultado: {num1} + {num2} = {resultado}")
            elif opcion == '2':
                resultado = resta(num1, num2)
                print(f"Resultado: {num1} - {num2} = {resultado}")
            elif opcion == '3':
                resultado = multiplicacion(num1, num2)
                print(f"Resultado: {num1} * {num2} = {resultado}")
            elif opcion == '4':
                resultado = division(num1, num2)
                print(f"Resultado: {num1} / {num2} = {resultado}")
        # 4. Implementar try-except para manejar errores
        except ValueError as e:
            print(e)
            print("Por favor, introduce números válidos.")
        except Exception as e:
            print(f"Ocurrió un error inesperado: {e}")
    else:
        print("Opción no válida. Por favor, selecciona un número del 1 al 5.")
    # 5. Usar continue para volver al menú principal después de cada operación (implícito en el bucle while)
```

```python
# ============================================
#        LISTA DE TAREAS - PYTHON
# ============================================

tareas = []

def agregar_tarea(tarea):
    tareas.append(tarea)
    print(f"\n✅ Tarea '{tarea}' agregada exitosamente.")

def eliminar_tarea(numero):
    try:
        tarea_eliminada = tareas.pop(numero - 1)
        print(f"\n🗑️  Tarea '{tarea_eliminada}' eliminada exitosamente.")
    except IndexError:
        print("\n❌ Error: El número de tarea no existe.")
    except ValueError:
        print("\n❌ Error: Ingresa un número válido.")

def mostrar_tareas():
    if not tareas:
        print("\n📋 No hay tareas pendientes.")
    else:
        print("\n📋 LISTA DE TAREAS:")
        print("-" * 30)
        for i, tarea in enumerate(tareas, start=1):
            print(f"  {i}. {tarea}")
        print("-" * 30)

# ============================================
#              MENÚ PRINCIPAL
# ============================================

print("=" * 40)
print("     GESTOR DE LISTA DE TAREAS")
print("=" * 40)

while True:
    print("\n📌 MENÚ PRINCIPAL:")
    print("  1. Agregar tarea")
    print("  2. Eliminar tarea")
    print("  3. Mostrar tareas")
    print("  4. Salir")

    opcion = input("\nSelecciona una opción (1-4): ").strip()

    if opcion == "1":
        nueva_tarea = input("📝 Escribe la nueva tarea: ").strip()
        if nueva_tarea:
            agregar_tarea(nueva_tarea)
        else:
            print("\n⚠️  La tarea no puede estar vacía.")

    elif opcion == "2":
        mostrar_tareas()
        if tareas:
            try:
                numero = int(input("🔢 Ingresa el número de tarea a eliminar: "))
                eliminar_tarea(numero)
            except ValueError:
                print("\n❌ Error: Debes ingresar un número entero.")

    elif opcion == "3":
        mostrar_tareas()

    elif opcion == "4":
        print("\n👋 ¡Hasta luego! Cerrando el programa...")
        break

    else:
        print("\n⚠️  Opción no válida. Por favor selecciona entre 1 y 4.")
```
